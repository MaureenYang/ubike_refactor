```python
import os,sys
import pymongo
import pandas as pd
import datetime
import parser
```

```python
p = parser.data_parser()

ubikedb = p.connect2db()
ubikecol_s = ubikedb["new_ubike_col"]
ubikecol = ubikedb["ubike_data_min"]

srcfilepath = "raw_data/youbike_data/"
filesinpath = os.listdir(srcfilepath)
```

```python
def db_iterate_find(col,query,batchsize=1000,order='asc',hint=None):
    #print(query)
    _query = query
    _re_df = pd.DataFrame()
    
    _sort =  ('_id', -1)
    if order == "asc":
            _sort =  ('_id', 1)
    #print("start")
    while True:
        
        if hint:
            queryResults = col.find(_query).sort([_sort]).limit(batchsize).hint(hint)
        else:
            queryResults = col.find(_query).sort([_sort]).limit(batchsize)
        
        df = pd.DataFrame(list(queryResults))
        _re_df = _re_df.append(df)
        query_cnt = len(df)

        #print('query_cnt:',query_cnt)
        
        if query_cnt == 0:
            break
        
        if query_cnt < batchsize:
            break
        
        last_id = df['_id'].iloc[-1]
        _query['_id'] = {
            '$lt': last_id
        }

        if order == "asc":
              _query['_id'] = {
                  '$gt': last_id
            }
    #print("end")

    #_re_df.to_csv("find_result.csv")
```

```python
#d = ubikecol.find({'sno':1})
start = datetime.datetime(2018, 7, 31, 0, 0, 0)
start_s = datetime.datetime(2021, 3, 31, 0, 0, 0)
tdelta = datetime.timedelta(hours=12)
#end = datetime.datetime(2021, 4, 1, 0, 0, 0)
end = start + tdelta
end_s = start_s + tdelta
#d = ubikecol.find( {'sno': 1, 'mday': {'$lt': end, '$gte': start}}).limit(3)
#query = {'sno': 1, 'mday': {'$lt': end, '$gte': start}}
#query = {'mday': {'$lt': end, '$gte': start}}

print("Small Query:")
query = {'mday': {'$lt': end_s, '$gte': start_s}}
batchsize_list = [50,100,200,500,1000,10000]
for i in batchsize_list:
    print('batch size:',i)
    db_iterate_find(ubikecol_s,query,batchsize=i)
```

```python
print("Big Query:")
query = {'mday': {'$lt': end, '$gte': start}}
batchsize_list = [100,200,500,1000,5000,10000]
for i in batchsize_list:
    print('batch size:',i)
    %timeit db_iterate_find(ubikecol,query,batchsize=i)
```
